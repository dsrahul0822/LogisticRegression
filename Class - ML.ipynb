```python
import pandas as pd
import numpy as np

def preprocessing_pipeline(df, target_col=None):
    """
    Complete preprocessing pipeline:
    1. Fills missing values in numerical columns with median
    2. Fills missing values in categorical columns with mode
    3. Applies one-hot encoding to categorical columns
    
    Parameters:
    -----------
    df : pandas DataFrame
        Input dataframe to preprocess
    target_col : str, optional
        Name of target column to exclude from preprocessing
    
    Returns:
    --------
    df_processed : pandas DataFrame
        Preprocessed dataframe
    """
    
    # Make a copy to avoid modifying original dataframe
    df = df.copy()
    
    # Separate target column if provided
    if target_col is not None and target_col in df.columns:
        target = df[target_col]
        df = df.drop(columns=[target_col])
    
    # ---------- STEP 1: Identify column types ----------
    numerical_cols = df.select_dtypes(include=['int64', 'float64', 'int32', 'float32']).columns.tolist()
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
    
    print("=" * 60)
    print("COLUMN IDENTIFICATION")
    print("=" * 60)
    print(f"Numerical columns ({len(numerical_cols)}): {numerical_cols}")
    print(f"Categorical columns ({len(categorical_cols)}): {categorical_cols}")
    
    # ---------- STEP 2: Missing value treatment for NUMERICAL columns (MEDIAN) ----------
    print("\n" + "=" * 60)
    print("MISSING VALUE TREATMENT - NUMERICAL (MEDIAN)")
    print("=" * 60)
    for col in numerical_cols:
        missing_count = df[col].isnull().sum()
        if missing_count > 0:
            median_value = df[col].median()
            df[col] = df[col].fillna(median_value)
            print(f"  {col}: filled {missing_count} missing values with median = {median_value}")
        else:
            print(f"  {col}: no missing values")
    
    # ---------- STEP 3: Missing value treatment for CATEGORICAL columns (MODE) ----------
    print("\n" + "=" * 60)
    print("MISSING VALUE TREATMENT - CATEGORICAL (MODE)")
    print("=" * 60)
    for col in categorical_cols:
        missing_count = df[col].isnull().sum()
        if missing_count > 0:
            mode_value = df[col].mode()[0]   # mode() returns a Series, take first value
            df[col] = df[col].fillna(mode_value)
            print(f"  {col}: filled {missing_count} missing values with mode = '{mode_value}'")
        else:
            print(f"  {col}: no missing values")
    
    # ---------- STEP 4: One-Hot Encoding for categorical columns ----------
    print("\n" + "=" * 60)
    print("ONE-HOT ENCODING")
    print("=" * 60)
    print(f"Applying get_dummies on: {categorical_cols}")
    print(f"Settings -> dtype=int, drop_first=True")
    
    df = pd.get_dummies(df, columns=categorical_cols, dtype=int, drop_first=True)
    
    print(f"Shape after encoding: {df.shape}")
    
    # ---------- STEP 5: Add target column back if it was separated ----------
    if target_col is not None:
        df[target_col] = target.values
    
    print("\n" + "=" * 60)
    print("PIPELINE COMPLETED SUCCESSFULLY")
    print("=" * 60)
    print(f"Final shape: {df.shape}")
    print(f"Remaining missing values: {df.isnull().sum().sum()}")
    
    return df


# ==================== USAGE EXAMPLE ====================
if __name__ == "__main__":
    
    # Example: Load your dataset
    # df = pd.read_csv('your_data.csv')
    
    # Sample data for demonstratio
    
    print("ORIGINAL DATAFRAME:")
    print(output)
    print(f"\nShape: {output.shape}")
    print(f"Missing values:\n{df.isnull().sum()}\n")
    
    # Run the pipeline
    df_clean = preprocessing_pipeline(output)
    
    print("\nFINAL PROCESSED DATAFRAME:")
    print(df_clean)
```
